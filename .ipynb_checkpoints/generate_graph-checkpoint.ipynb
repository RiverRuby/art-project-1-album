# Generating Graphs

Edges between (Art A, Alb A)  and (Art B, Alb B)
  1. Art A == Art B (same artist)
  2. Art A featured in Alb B || Art B featured in Alb A
  3. Exists song x in Alb A, song y in Alb B s.t. x.name == y.name

```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import json
```

```python
scope = "user-library-read"
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id='d33d36b61fc64052bd94e69a9ea1f9fe',
    client_secret='3d4c68cce8d3427abf9c2ebb1f04b40b',
    redirect_uri='http://localhost:8888/',
    scope=scope
))

offset = 0
all_albums = []

while offset < 127:
    results = sp.current_user_saved_albums(50, offset)
    all_albums.extend(results['items'])
    offset += 50
```

```python
# removing certain albums
bl_ind = -1
for i in range(len(all_albums)):
    if all_albums[i]['album']['name'] == 'ESCAPE PLAN / MAFIA':
        bl_ind = i
        break

all_albums.pop(bl_ind)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
for album in all_albums:
    G.add_node(album['album']['name'])
```

```python
# get edges between albums by the same artist
artists = {}
for album in all_albums:
    for artist in album['album']['artists']:
        name = artist["name"]
        if name not in artists:
            artists[name] = []
        artists[name].append((album['album']['release_date'], album['album']))

for name in artists:
    artists[name] = sorted(artists[name])
    for i in range(len(artists[name]) - 1):
        G.add_edge(artists[name][i][1]['name'], artists[name][i+1][1]['name'])
```

```python
nx.draw(G, with_labels=True)
```

```python
from pyvis.network import Network

nt = Network(width=1000, height=500, notebook=True)
nt.from_nx(G)
nt.show("ex.html")
```

```python
from datetime import datetime
import json

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```

```python
# if artist a1 featured on album, add edge between that album and closest album in a1's discography
for idx, album in enumerate(all_albums):
    main_artists = set([])
    for artist in album['album']['artists']:
        main_artists.add(artist["name"])
    
    featured_artists = set([])
    for track in album['album']['tracks']['items']:
        for artist in track['artists']:
            if artist['name'] not in main_artists:
                featured_artists.add(artist['name'])
    
    for fname in featured_artists:
        if fname in artists:
            mini = 365*3
            min_ind = -1
            for i in range(0, len(artists[fname])):
                if days_between(artists[fname][i][0], album['album']['release_date']) < mini:
                    mini = days_between(artists[fname][i][0], album['album']['release_date'])
                    min_ind = i
            if min_ind != -1:
                G.add_edge(album['album']['name'], artists[fname][i][1]['name'], color='red')
                print(album['album']['name'], artists[fname][i][1]['name'])
```

```python
nt = Network(width=1000, height=500, notebook=True)
nt.from_nx(G)
nt.show("ex.html")
```
